```python
def neuron_w(input_count):
  weights = np.zeros(input_count + 1)
  for i in range(1,(input_count+1)):
    weights[i] = np.random.uniform(-1.0, 1.0)
    return weights
n_w = [neuron_w(2), neuron_w(2), neuron_w(2)] 
n_y = [0, 0, 0]
n_error = [0, 0, 0]
```
